```python
import pandas as pd
import time
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("Car_Final.csv")
x=df.iloc[:,[0,1,2,3]].values
df
```

```python
scaler=MinMaxScaler()
df_1=scaler.fit_transform(df)
df_2=pd.DataFrame(df_1,columns=df.columns)
df_2
```

```python
test=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    test.append(kmeans.inertia_)
```

```python
Start=time.time()
kmeans=KMeans(n_clusters=4,init='k-means++',max_iter=300,n_init=10,random_state=5)
kmeans.fit(x)
y_kmeans=kmeans.fit_predict(x)
print("分群結果:")
print(y_kmeans)
print("---------------------")
End=time.time()
print("運算時間: ",End-Start)
```

```python
kmeans_add=df_2
kmeans_add["Cluster"]=y_kmeans
kmeans_add
```

```python
kmeans_results=kmeans_add.groupby("Cluster").sum()
kmeans_results
```

```python
purityList=[]
total=0
for i in range(3):
    purity=kmeans_results.loc[i].max()/kmeans_results.loc[i].sum()
    purityList.append(purity)
    total=total+kmeans_results.loc[i].sum()
```

```python
print("純度:\n",purityList)
print("總和:\n",total)
```

```python
Total_purity=0
for i in range(len(purityList)):
    Total_purity=Total_purity+purityList[i]*(kmeans_results.loc[i].sum()/total)
print("總純度:\n",Total_purity)
```
