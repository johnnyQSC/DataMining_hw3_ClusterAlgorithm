```python
import pandas as pd
import time
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN
```

```python
df=pd.read_csv("Car_Final.csv")
df
```

```python
scaler=MinMaxScaler()
df_1=scaler.fit_transform(df)
df_2=pd.DataFrame(df_1,columns=df.columns)
```

```python
Start = time.time()
DBSCAN = DBSCAN(eps=0.6,min_samples=15).fit(df_2)
DBSCAN_label=DBSCAN.fit_predict(df_2)
print("分群結果:")
print(DBSCAN_label)
print("---------------------")
End = time.time()
print("運算時間: ",End-Start)
```

```python
len(DBSCAN_label)
```

```python
DBSCAN_add = df_2
DBSCAN_add["Cluster"] = DBSCAN_label
DBSCAN_add
```

```python
DBSCAN_results = DBSCAN_add.groupby("Cluster").sum()
DBSCAN_results
```

```python
purityList=[]
total=0
for i in range(1):
    purity = DBSCAN_results.loc[i].max()/DBSCAN_results.loc[i].sum()
    purityList.append(purity)
    total = total + DBSCAN_results.loc[i].sum()

print("純度:\n",purityList)
print("總和:\n",total)
```

```python
Total_purity=0
for i in range(len(purityList)):
    Total_purity = Total_purity+purityList[i]*(DBSCAN_results.loc[i].sum()/total)
print("總純度:\n",Total_purity)
```
