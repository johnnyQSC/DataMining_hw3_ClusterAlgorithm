```python
import pandas as pd
```

```python
df=pd.read_csv("car.csv")
df
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df_1 = df
df_1['buying']=labelencoder.fit_transform(df['buying'])
df_1['maint']=labelencoder.fit_transform(df['maint'])
df_1['doors']=labelencoder.fit_transform(df['doors'])
df_1['persons']=labelencoder.fit_transform(df['persons'])
df_1['lug_boot']=labelencoder.fit_transform(df['lug_boot'])
df_1['safety']=labelencoder.fit_transform(df['safety'])
df_1['Class Values']=labelencoder.fit_transform(df['Class Values'])
df_1
```

```python
df_1.to_csv("Car_Final.csv",index=False,encoding="utf-8-sig")
```
